5. Maktabda baholash tizimi quyidagi qoidalarga ega:

- a. 25 dan past - F
- b. 25 dan 45 gacha - E
- c. 45 dan 50 gacha - D
- d. 50 dan 60 gacha - C
- e. 60 dan 80 gacha - B
- f. 80 dan yuqori - A

Foydalanuvchidan belgilar kiritishini va tegishli bahoni chop etishini so'rang.

```python
# javobni shu yerga kiriting...
ball = int(input("ballni kiriting"))
if ball < 25 and 0 <= ball :
   print("Siznining darajangiz F")
elif 25 <= ball and ball < 45 : 
   print("Sizning darajangiz E")
elif 45 <= ball and ball < 50 :
    print("Sizning darajangiz D")
elif 50 <= ball and ball < 60 :
    print("Sizning darajangiz C")
elif 60 <= ball and ball < 80 :
    print("Sizning darajangiz B")
elif 80 <= ball :
    print("Sizning darajangiz A")
else:
    print("Natija aniqlanmadi")
```

6. Foydalanuvchi tomonidan 3 kishining yoshini kiriting. Ular orasida eng keksa va eng kichigini aniqlang.

```python
# javobni shu yerga kiriting...
a = int(input("1-yoshingizni kiriting"))
b = int(input("2-yoshingizni kiriting"))
c = int(input("3-yoshingizni yozing"))
if a > b and a > c and b > c :
    print("3-kiritilgan yosh eng kichigi")
    print("1-kiritilgan yosh eng kattasi")
elif a > b and a > c and c > b :
    print("2-kiritilgan yosh eng kichigi")
    print("1-kiritilgan yosh eng kattasi")
elif b > a and b > c and a > c :
    print("3-kiritilgan yosh eng kichigi")
    print("2-kiritilgan yosh eng kattasi")
elif b > a and b > c and c > a :
    print("1-kiritilgan yosh eng kichigi")
    print("2-kiritilgan yosh eng kattasi")
elif c > a and c > b and a > b :
    print("2-kiritilgan yosh eng kichigi")
    print("3-kiritilgan yosh eng kattasi")
elif c > a and c > b and b > a :
    print("1-kiritilgan yosh eng kichigi")
    print("3-kiritilgan yosh eng kattasi")
else:
    Print("Yoshlar teng bo`lib qoldi")
```

7. Bir yil kabisa yili yoki yo'qligini tekshirish uchun dastur yozing.
