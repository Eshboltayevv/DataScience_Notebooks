8. `uybor_second.xlsx` faylini o'qing va `uy_df2` deb dataframe ko'rinishida saqlab oling. Keying `uy_df` va `uy_df2` larni birlashtiring va `full_df` deb saqlang(Maslahat concat ishlating.)

```python
# javobni shu yerga kiriting...
import pandas as pd
uy_df = pd.read_excel("uybor_first.xlsx")
uy_df2 = pd.read_excel("uybor_second.xlsx")
full_df = pd.concat([uy_df , uy_df2])
full_df.head()
```

9. `full_df` dataframedan tumani Yakkasaroy bo'lgan va narxi 25000 dan past bo'lgan  uylarni alohida dataframe qling.Shu dataframeni `Yakkasaroy.xlsx` deb saqlab qoying.

```python

yakkasaroy_df = full_df[(full_df['district'] == 'Yakkasaroy') & (full_df['price'] < 25000)]

yakkasaroy_df.to_excel("Yakkasaroy.xlsx", index=False)

yakkasaroy_df.head()
```
