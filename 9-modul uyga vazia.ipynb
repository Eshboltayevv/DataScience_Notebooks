````python
# javobni shu yerga kiriting...
def calc(a, b, amal):
    if amal == "+":
        return a + b
    elif amal == "-":
        return a - b
    elif amal == "*":
        return a * b
    elif amal == "/":
        if b != 0:
            return a / b
        else:
            return "0 ga bo`lish mumkin emas"
    else:
        return "Noto`g`ri amal kiritildi"

print(calc(3, 0, "/"))  
````

6. Satrni qabul qiladigan, katta va kichik harflar sonini hisoblaydigan funksiya yozing.

Misol uchun : 
``` "O'zbekiston Respublikasi poytaxti Toshkent shahri"  ```
``` 'sh', 'ch', 'ng', 'Sh', 'Ch' ```  inobatga olinsin.


````python
# javobni shu yerga kiriting...
def harf_soni(matn):
    kichik = 0
    katta = 0
    
    maxsus_kichik = ['sh', 'ch', 'ng']
    maxsus_katta = ['Sh', 'Ch']
    
    i = 0
    while i < len(matn):
        if matn[i].isalpha():

            if matn[i:i+2] in maxsus_kichik:
                kichik += 1
                i += 2
                continue
            elif matn[i:i+2] in maxsus_katta:
                katta += 1
                i += 2
                continue
            
    
            if matn[i].islower():
                kichik += 1
            elif matn[i].isupper():
                katta += 1
        i += 1
    
    return katta, kichik
matn = "O'zbekiston Respublikasi poytaxti Toshkent shahri"
katta, kichik = harf_soni(matn)
print("Katta harflar:", katta)
print("Kichik harflar:", kichik)
````

7. Berilgan o'nlik sonini ikkilik sanoq sistemasiga o'tkazuvchi funksiya yozing. Funksiya n o'nlik sonini argument sifatida qabul qiladi va uni ikkilik sanoq sistemasiga o'tkazadi.

````python
# javobni shu yerga kiriting...
def Onlik_son(n):
    if n == 0:
        return "0"
    
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary 
        n = n // 2                  
    return binary

son = 17
print(Onlik_son(son)) 
````

8. Berilgan ikkilik sonni o'nlik sanoq sistemasiga o'tkazuvchi funksiya yozing.


````python
# javobni shu yerga kiriting...
def Onlik_son(binar):
    decimal = 0
    binar = str(binar)  
    uzunlik = len(binar)
    
    for i in range(uzunlik):
        
        decimal += int(binar[i]) * (2 ** (uzunlik - i - 1))
    
    return decimal

ikkilik_son = "1101"
print(Onlik_son(ikkilik_son))
````

9. Berilgan ikkita butun sonning eng kichik umumiy bo'luvchisini (EKUB) topuvchi funksiya yozing. Funksiya a va b sonlarini argument sifatida qabul qiladi va ularning EKUBini qaytaradi.

````python
# javobni shu yerga kiriting...
def ekub(a, b):
    
    while b != 0:
        a, b = b, a % b
    return a


x = 24
y = 36
print(ekub(x, y))  
````

10. Berilgan ikkita butun sonning eng kichik umumiy karralisini (EKUK) topuvchi funksiya yozing. Funksiya a va b sonlarini argument sifatida qabul qiladi va ularning EKUKini qaytaradi.

````python
# javobni shu yerga kiriting...
def ekub(a, b):
    
    while b != 0:
        a, b = b, a % b
    return a

def ekuk(a, b):
    return abs(a * b) // ekub(a, b)


x = 12
y = 18
print(ekuk(x, y))  
````

11. Berilgan musbat butun sonning kvadrat ildizini topuvchi funksiya yozing. 
Kutubxonalardan foydalanish mumkun emas.

````python
# javobni shu yerga kiriting...
def kvadrat_ildiz(n):
    if n == 0 or n == 1:
        return n  

    past = 0
    yuqori = n
    epsilon = 1e-6  

    while yuqori - past > epsilon:
        mid = (past + yuqori) / 2
        if mid * mid > n:
            yuqori = mid
        else:
            past = mid

    return past


son = 16
print(kvadrat_ildiz(son))
````
