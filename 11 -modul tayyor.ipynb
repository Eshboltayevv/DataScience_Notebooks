2. `matn_chastota.txt` ichidagi ma'lumotlarni lug'at ko'rinishiga o'tkazing.

3. Hosil bo'lgan lug'atdan eng ko'p uchragan so'zni toping.


````python
# javobni shu yerga kiriting...
let = {}

with open("matn_chastota.txt", encoding="utf-8") as f:
    for line in f:
        count, word = line.split()
        let[word] = int(count)

eng_kop_soz = max(let, key=let.get)

print("Eng ko‘p uchragan so‘z:", eng_kop_soz)
print("Takrorlanish soni:", let[eng_kop_soz])
````

4. Hosil bo'lgan lug'atda bir marta uchragan so'zlarni hammasini ajratib oling. 

````python
# javobni shu yerga kiriting...
bir_marta = []

for soz, son in let.items():
    if son == 1:
        bir_marta.append(soz)

print(bir_marta)
````

5. `matn_chastota.txt` faylni o'qing va 'ch' harfi bilan boshlangan barcha so'zlarni yig'indisini hisoblang.

````python
# javobni shu yerga kiriting...
yigindi = 0

with open("matn_chastota.txt", encoding="utf-8") as f:
    for line in f:
        count, word = line.split()
        count = int(count)

        if word.startswith("ch"):
            yigindi += count

print("`ch` bilan boshlangan so‘zlar yig‘indisi:", yigindi)
````
